## Open CV Face Detection


#### Import Open CV & Assign HaarCascade

```python

import cv2
face_detector=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

#### Read Image and Convert to GrayScale

```python
image=cv2.imread("faces.jpg")
```

```python
from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
GR_img = cv2.cvtColor(RGB_img,cv2.COLOR_BGR2GRAY)
plt.imshow(GR_img)
```

#### Detect coordinates of Faces

```python
results = face_detector.detectMultiScale(GR_img, scaleFactor=1.5,minNeighbors=5,minSize=(30, 30), flags=cv2.CASCADE_SCALE_IMAGE)
print(results)
```

```python
image=cv2.imread("faces.jpg")
for(x,y,w,h) in results:
    cv2.rectangle(image,(x,y),(x+w,y+h),(250,0,0),2)

RGB_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```
